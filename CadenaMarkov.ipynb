```julia
function CadenaMarkov(network)
    matriz_adyacencia = full(adjacency_matrix(network))
    matriz_adyacencia = Array{Float64,2}[matriz_adyacencia][1]
    n = length(matriz_adyacencia[1, :])
    for i = 1:n
        s = sum(matriz_adyacencia[i, :])
        #println(s)
        matriz_adyacencia[i, :] = (1./s)*matriz_adyacencia[i, :]
    end
    return matriz_adyacencia
end
```

```julia
MC_Normal = CadenaMarkov(RedMetroNormal);
MC_Normal_Estacionaria = MC_Normal^50;
```

```julia
imshow(MC_Normal_Estacionaria)
title("Convergencia de la Cadena de Markov")
colorbar()
```
